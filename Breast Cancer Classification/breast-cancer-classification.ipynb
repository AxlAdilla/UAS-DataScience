```python
import pandas as pd
import numpy as np
```

```python
dataset = pd.read_csv('data.csv', index_col='id')
dataset.head()
```

```python
dataset.drop(dataset.columns[-1], axis=1, inplace=True)
```

```python
dataset.describe()
```

```python
dataset.isna().sum()
```

```python
dataset.info()
```

```python
count_class_b, count_class_m = dataset['diagnosis'].value_counts()
```

```python
# Divide by class
dataset_class_b = dataset[dataset['diagnosis'] == 'B']
dataset_class_m = dataset[dataset['diagnosis'] == 'M']
```

```python
dataset_class_m_over = dataset_class_m.sample(count_class_b, replace=True)
```

```python
dataset = pd.concat([dataset_class_b, dataset_class_m_over])
dataset['diagnosis'].value_counts()
```

```python
dataset_y = dataset['diagnosis']
dataset_y
```

```python
dataset_x = dataset.drop('diagnosis', axis=1)
dataset_x
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(dataset_x, dataset_y, random_state=0, test_size=0.7 )
```

```python
x_train
```

```python
def diagnosa_to_label(diagnosa):
    if diagnosa == 'M':
        return 1
    else:
        return 0

y_train_label = y_train.apply(diagnosa_to_label)
y_train_label
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()

x_train_scaled = scaler.fit_transform(x_train)
x_train_scaled
```

```python
from sklearn.feature_selection import SequentialFeatureSelector
from sklearn.metrics import accuracy_score
from sklearn.svm import SVC
import time

clf = SVC()
total_accuracy = []
total_n_fitur = []
total_time = []
with open(f'report_fitur.csv', 'w+') as f:
    f.write('Jumlah Fitur,Akurasi,time,Kolom Digunakan\n')

for n_fitur in range(1, len(x_train.columns)):
    sfs = SequentialFeatureSelector(clf, n_features_to_select=n_fitur, direction='forward')
    sfs.fit(x_train_scaled, y_train_label)
    new_x_train_scaled = sfs.transform(x_train_scaled)

    y_test_label = y_test.apply(diagnosa_to_label)
    x_test_scaled = scaler.transform(x_test)
    new_x_test_scaled = sfs.transform(x_test_scaled)
    
    start_train = time.process_time()
    clf.fit(new_x_train_scaled, y_train_label)  
    y_pred = clf.predict(new_x_test_scaled)
    end_train = time.process_time() - start_train
    
    accuracy_model = accuracy_score(y_test_label, y_pred)
    total_accuracy.append(accuracy_model)
    total_n_fitur.append(n_fitur)
    total_time.append(end_train)
    
    with open(f'report_fitur.csv', 'a+') as f:
        f.write(f'{n_fitur},{accuracy_model},{end_train:.12f},{"; ".join(list(x_train.columns[sfs.get_support()]))} \n')
```

```python
import matplotlib.pyplot as plt

plt.plot(total_n_fitur, total_accuracy)
plt.xlabel('n-Fitur')
plt.ylabel('Accuracy')
#plt.show()
plt.savefig('Grafik Accuracy.png')
plt.clf()

plt.plot(total_n_fitur, total_time)
plt.xlabel('n-Fitur')
plt.ylabel('Accuracy')
#plt.show()
plt.savefig('Grafik Time.png')
```
